```python
import cobra
```

```python
model = cobra.Model("single_pathway")
model.add_metabolites([cobra.Metabolite("A", compartment="e"), cobra.Metabolite("B", compartment="e")])

model.add_reaction(cobra.Reaction("import_A"))
model.reactions.import_A.add_metabolites({"A": 1})

model.add_reaction(cobra.Reaction("A_to_B"))
model.reactions.import_A.add_metabolites({"A": -1, "B": 1})

model.add_reaction(cobra.Reaction("export_B"))
model.reactions.import_A.add_metabolites({"B": -1})

cobra.io.write_sbml_model(model, "models/single_pathway.xml")
```

```python
model = cobra.Model("branching_pathways")
model.add_metabolites(
    [cobra.Metabolite("A", compartment="e"),
     cobra.Metabolite("Right", compartment="e"), 
     cobra.Metabolite("Left", compartment="e")
     ])

model.add_reaction(cobra.Reaction("import_A"))
model.reactions.import_A.add_metabolites({"A": 1})

model.add_reaction(cobra.Reaction("A_to_Right"))
model.reactions.import_A.add_metabolites({"A": -1, "Right": 1})

model.add_reaction(cobra.Reaction("export_Right"))
model.reactions.import_A.add_metabolites({"Right": -1})

model.add_reaction(cobra.Reaction("A_to_Left"))
model.reactions.import_A.add_metabolites({"A": -1, "Left": 1})

model.add_reaction(cobra.Reaction("export_Left"))
model.reactions.import_A.add_metabolites({"Left": -1})

cobra.io.write_sbml_model(model, "models/branching_pathway.xml")
```

```python
model = cobra.Model("roundabout")
model.add_metabolites(
    [cobra.Metabolite("A", compartment="e"),
     cobra.Metabolite("B", compartment="e"), 
     cobra.Metabolite("C", compartment="e")
     ])

model.add_reaction(cobra.Reaction("import_A"))
model.reactions.import_A.add_metabolites({"A": 1})

model.add_reaction(cobra.Reaction("import_B"))
model.reactions.import_A.add_metabolites({"B": 1})

model.add_reaction(cobra.Reaction("import_C"))
model.reactions.import_A.add_metabolites({"C": 1})

model.add_reaction(cobra.Reaction("A_to_B"))
model.reactions.import_A.add_metabolites({"A": -1, "B": 1})

model.add_reaction(cobra.Reaction("B_to_C"))
model.reactions.import_A.add_metabolites({"B": -1, "C": 1})

model.add_reaction(cobra.Reaction("C_to_A"))
model.reactions.import_A.add_metabolites({"C": -1, "A": 1})

cobra.io.write_sbml_model(model, "models/roundabout.xml")
```
